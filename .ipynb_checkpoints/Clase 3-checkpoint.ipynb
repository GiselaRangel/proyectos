```python
from pylab import * 
%matplotlib inline
```

```python
def func(x):
    y= x**2
    return y
```

```python
x = linspace (0,2)
plot (x,func(x))
```

```python
def derivador(func,x,h):
    y = (func(x+h) - func(x))/h
    return y
```

```python
plot (x,func(x))
plot (x,derivador(func,x,0.01))
plot (x,2*x)
```

```python
plot (x,func(x))
plot (x,derivador(func,x,0.5))
plot (x,2*x)
```

```python
def mi_ec_dif(y,x):
    return y
```

```python
def euler (func,p_ini,h,x_f):
    pasos = arange(p_ini[0],x_f+h,h)
    y = zeros (len(pasos))
    y[0] = p_ini[1]
    for i in range(len(pasos)-1):
        y[i+1] = y[i] + h*func(y[i],pasos[i])
    solucion = [pasos, y]
    return solucion
```

```python
arange (1,4,0.2)
```

```python
sol = euler(mi_ec_dif,[0,1],0.01,200)
plot (sol[0],sol[1],sol[0],exp(sol[0]))
xlim(170,200)
```

```python
def ec_2(y,x):
    return -x*y
```

```python
sol = euler(ec_2,[0,1],0.5,10)
plot(sol[0],sol[1],sol[0],exp((-sol[0]**2)/2))
```
